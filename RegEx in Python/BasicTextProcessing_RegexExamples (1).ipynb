```python
import re
text= "Begin the the hello hello"
print(re.match(r'\b([a-zA-Z]+)\s+\1\b',text))##Determines if the RE matches at the beginning of the string, returns None or Match object-->1rst match ONLY
print(re.search(r'\b([a-zA-Z]+)\s+\1\b',text))##Scan through the string looking for any location where RE matches, returns None or Match object-->1rst Match ONLY
print(re.findall(r'\b([a-zA-Z]+)\s+\1\b',text))##Find all substrings where RE matches, returns them as list-->All Matches
[x.group() for x in re.finditer(r'\b([a-zA-Z]+)\s+\1\b',text)]##Find all substrings where RE matches, returns them as iterator-->All Matches
                                                              ##group() is a function of match object that returns entire match
```

```python
import re
string_original="hi I am 2   3 4 living at 12 6 street 98 on road 1 and have 1- 2 3 7cats"

##re.sub(pattern, repl, string, max=0)
##This method replaces all occurrences of the RE pattern in string with repl, substituting all occurrences unless max is provided. This method returns modified string.
string_new = re.sub(r"(?<=\d)\s+(?=\d)", '', string_original)

print(string_new)
```

```python
import re
text="010010101 END"
print(re.match(r'(0*1+){4,}0*',text))
print(re.search(r'(0*1+){4,}0*',text))
print(re.findall(r'(0*1+){4,}0*',text))
print(re.findall(r'(?:0*1+){4,}0*',text))##Using non-capturing group to print the whole match
print([x.group() for x in re.finditer(r'(0*1+){4,}0*',text)])

print("With BEG at string")
text="BEG 010010101 END"
print(re.match(r'(0*1+){4,}0*',text))
print(re.search(r'(0*1+){4,}0*',text))
print(re.findall(r'(0*1+){4,}0*',text))
print(re.findall(r'(?:0*1+){4,}0*',text))
[x.group() for x in re.finditer(r'(0*1+){4,}0*',text)]
```

```python
'''
we can use raw string r"string", in raw strings no escaping is required as escape sequences like \n \t are not proceed.

'''
print(re.search(r"C:\\Windows\\Systems",txt))
print("hello\tworld")
print(r"hello\tworld")
```
